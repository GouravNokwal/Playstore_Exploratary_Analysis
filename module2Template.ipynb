## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# playstore


# 1. Drop duplicate rows
ps = playstore.drop_duplicates()

# 2. Handle missing values (drop rows where critical info is missing)
ps = ps.dropna(subset=['Rating', 'Category', 'Installs', 'Price'])

# 3. Clean 'Installs' column (remove + , and convert safely to int)
ps['Installs'] = ps['Installs'].str.replace('[+,]', '', regex=True)
ps['Installs'] = pd.to_numeric(ps['Installs'], errors='coerce')
ps = ps.dropna(subset=['Installs'])
ps['Installs'] = ps['Installs'].astype(int)

# 4. Clean 'Price' column (remove $ and convert safely to float)
ps['Price'] = ps['Price'].str.replace('$', '', regex=True)
ps['Price'] = pd.to_numeric(ps['Price'], errors='coerce').fillna(0)

# 5. Clean 'Size' column
def clean_size(x):
    if isinstance(x, str):
        if 'M' in x:
            return float(x.replace('M', '')) * 1e6
        elif 'k' in x:
            return float(x.replace('k', '')) * 1e3
        elif x.strip() == 'Varies with device':
            return np.nan
    return x

ps['Size'] = ps['Size'].apply(clean_size)
ps['Size'] = pd.to_numeric(ps['Size'], errors='coerce')

# 6. Clean 'Reviews' column (convert to int)
ps['Reviews'] = pd.to_numeric(ps['Reviews'], errors='coerce').fillna(0).astype(int)

# 7. Reset index
ps = ps.reset_index(drop=True)
````

````python
# Userreview
# 1. Drop duplicates
ur = userreview.drop_duplicates()

# 2. Handle missing values (drop rows where Sentiment is missing)
ur = ur.dropna(subset=['Sentiment'])

# 3. Fill missing sentiment scores with 0
ur['Sentiment_Polarity'] = ur['Sentiment_Polarity'].fillna(0)
ur['Sentiment_Subjectivity'] = ur['Sentiment_Subjectivity'].fillna(0)

# 4. Reset index
ur = ur.reset_index(drop=True)
ur
````

### What all manipulations have you done and insights you found?

Play Store Data (ps)

Manipulations :

Removed duplicates to avoid bias in counts/averages.

Dropped rows with missing critical info (Rating, Category, Installs, Price).

Cleaned Installs column: removed + and ,, converted to integers.

Cleaned Price column: removed $, converted to float.

Converted Size column:

Values like "19M" → 19,000,000

Values like "400k" → 400,000

"Varies with device" → treated as NaN

Cleaned Reviews column: converted to integers.

Reset index after cleaning.


Insights :

Some apps had "Free" in price, but after cleaning, they are 0.0.

Install counts were inconsistent (10,000+, 1,000,000+) → now standard integers, so easier to aggregate.

App sizes varied drastically (from KB to 100+ MB) → can analyze relationship between size vs installs/ratings.

There were duplicate entries for same apps with slightly different details. Now, dataset is consistent.

User Reviews Data (ur)

Manipulations :

Removed duplicates to avoid repeated reviews.

Dropped missing Sentiment values (since those reviews cannot be used in sentiment analysis).

Filled missing sentiment polarity/subjectivity with 0.

Polarity = sentiment positivity/negativity score.

Subjectivity = how much personal opinion vs factual.

Reset index.

Insights:

Many reviews had no sentiment scores (missing), meaning they couldn’t be processed by NLP initially. After filling, we can still use them.

Distribution of sentiments (Positive, Neutral, Negative) can now be analyzed properly.

Some apps have a large imbalance of review types (e.g., a highly installed app might still have more negative reviews).
